```python
# Create youtube resource object
myYoutube = build('youtube',
                  'v3',
                  developerKey=API_KEY)
```

```python
# Retrieve youtube civdeo comments
video_response = myYoutube.commentThreads().list(
    part='snippet, replies',
    videoId='hBpepnlqaTo',
).execute()
```

```python
count = 0
# Iterate youtube comments
for item in video_response["items"]:
    count+=1
    print("Comment-", count, ": ")
    comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
    print(comment)
```

```python
# Extract userid, userimage, comment, and replies on comment
count = 0
# Iterate youtube comments
for item in video_response["items"]:
    count+=1
    print("Comment-", count, ": ")
    comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
    user = item["snippet"]["topLevelComment"]["snippet"]["authorDisplayName"]
    user_image = item["snippet"]["topLevelComment"]["snippet"]["authorProfileImageUrl"]
    print("User: ", user)
    print("User Image: ", user_image)
    print("Comment: ", comment)
    print("\n")
    print("Replies: ")
    try:
        replies = item["replies"]["comments"]
        for reply in replies:
            reply = reply["snippet"]["textDisplay"]
            print(reply)
    except:
        print("No replies")
    print("\n\n")
```
